```python
#Exercício 1 – Seleção avançada e filtros 
#Utilizando o DataFrame lido do ficheiro exercício_vendas.csv
import pandas as pd 
import numpy as np 
df_vendas = pd.read_csv("exercicio_vendas.csv")
display(df_vendas)

print("_________________________________________________________")
#a) Filtra todas as linhas em que a categoria é “Eletrónica” ou em que a quantidade vendida seja superior a 5
#df_vendas.query( Eletrónica' and 'Quantidade' > 5")

df_filtrado= df_vendas [(df_vendas['Categoria']== 'Eletrónica') & (df_vendas['Quantidade'] > 5 )]
display(df_filtrado)
print("__________________________________________________________")
#b) Mostra apenas as colunas “Produto”, “Categoria” e “Quantidade” dessas linhas filtradas

df_filtrado = df_filtrado[['Produto', 'Categoria', 'Quantidade']]
display(df_filtrado)
```

```python
#Exercício 2 – Tratamento de dados ausentes e duplicados 
#1. Conta, para cada coluna, quantos valores ausentes (NaN) existem no DataFrame original. 

qt_nulos= df_vendas.isna().sum()
print (qt_nulos)
print("________________________________________________________________")
#2. Preenche todos os valores ausentes na coluna “Preço” com a média dessa mesma coluna. 

media_preco= df_vendas['Preço'].mean()
print(media_preco)

df_vendas['Preço']=df_vendas['Preço'].fillna(media_preco)
display(df_vendas)
print("____________________________________________________________")

qt_media= df_vendas['Quantidade'].mean()
print(qt_media)
df_vendas['Quantidade']= df_vendas['Quantidade'].fillna(qt_media).astype(int)
display(df_vendas)
print("____________________________________________________________")


#3. Remove eventuais linhas duplicadas e mostra o novo DataFrame. 
df_vendas = df_vendas.drop_duplicates()
display(df_vendas)
```

```python
#Exercício 3 – Alteração de tipos de dados 
#1. Confere o tipo de dados da coluna “Quantidade”. 
df_vendas.info()
print(df_vendas['Quantidade'].dtypes)
print("___________________________________________________________________")

#2. Converte essa coluna para o tipo inteiro, se ainda não estiver. 

df_vendas['Quantidade']  = df_vendas['Quantidade'].astype(int)
display(df_vendas)
#3. Verifica novamente o tipo após a conversão. 



```

```python
#Exercício 4 – Análise integrada de vendas (exercício mais extenso e completo) 
#Realiza a seguinte análise, utilizando e combinando os métodos .loc[], .isna(), .fillna(), .drop_duplicates(), .astype(), 
#.groupby():

#4.1. Lê o ficheiro CSV e apresenta as primeiras 5 linhas. 
df_vendas_orig=pd.read_csv('exercicio_vendas_orig.csv')
display(df_vendas_orig)

display(df_vendas.head())

print("_________________________________________________________________________________________")

```

```python
#4.2. Conta os valores ausentes por coluna e preenche todos os ausentes na coluna “Quantidade” com a moda dessa 
#coluna e os ausentes da coluna «Preço» com a média dessa coluna.. 


valores_ausentes = df_vendas_orig.isna().sum()
print("Quantidade de valores ausentes")

print(valores_ausentes)
print("_________________________________________________________________________________________")

quant_media= df_vendas_orig['Quantidade'].mode()[0] # para retomar o primeiro valor da moda 
print(f" A quantidade média é {quant_media}")

df_vendas_orig["Quantidade"]=df_vendas_orig["Quantidade"].fillna(media_preco).astype(int)
display(df_vendas_orig)

print("_________________________________________________________________________________________")

preco_media=df_vendas_orig['Preço'].mean()
print(f" A média do preço é {media_preco}")

df_vendas_orig["Preço"]=df_vendas_orig["Preço"].fillna(media_preco).astype(float)
print(df_vendas_orig)
```

```python
##4.3.  Remove as linhas duplicadas

df_vendas_orig_sem_dup = df_vendas_orig.drop_duplicates() ##  a linha é removida, deixando de ter sequência,
display(df_vendas_orig_sem_dup)
```

```python
##4.4.Cria uma nova coluna “Valor_Total” que é o produto de “Quantidade” e “Preço”##


df_vendas_orig_sem_dup['Valor Total']=df_vendas_orig_sem_dup ['Quantidade'] * df_vendas_orig_sem_dup['Preço']
display(df_vendas_orig_sem_dup)
```

```python
##4.5.Confirma que a coluna “Valor_Total” está no tipo float.
df_vendas_orig_sem_dup.info()

print("________________________________________________")
print(df_vendas_orig_sem_dup['Valor Total'].dtype)
```

```python
## 4.6 Agrupa o DataFrame por “Categoria” e mostra:
# O número total de produtos vendidos por cada categoria (soma das quantidades)
# O valor total vendido por categoria (soma dos “Valor_Total”)

agrupar_ctg = df_vendas_orig_sem_dup.groupby('Categoria').agg(
    Produtos_vend=('Quantidade', 'sum'),
    valor_total_vend=('Valor Total', 'sum')
).reset_index()
display(agrupar_ctg)
```

```python
# 4.7
# Ordena o resultado final por valor total vendido, do maior para o menor.

Resultado_final= agrupar_ctg.sort_values(by= 'valor_total_vend', ascending=False)
display(Resultado_final)
```
