## Parte 4 — Estatística Descritiva - Utilize o dataframe limpo e tratado na Parte 3. 

```python
##) Calcule, para a coluna Salario: 
# Média, Mediana, Moda, Quartis (Q1, Q2, Q3), Desvio Padrão (amostra), Coeficiente de Variação. 

salario_media = round(df_tratado['Salario'].mean(), 2)
print(f"O salário médio é: {salario_media}")
print("______________________________________________________")

salario_mediana= round(df_tratado['Salario'].median(), 2)
print(f" A mediana do salário é : {salario_mediana}")
print("______________________________________________________")

salario_moda =df_tratado['Salario'].mode()[0]
print(f" A moda  da variável salário é : {salario_moda}")

quartis = round(df_tratado['Salario'].quantile([0.25, 0.5, 0.75]), 2)
Q1 = quartis[0.25]
Q2 = quartis[0.5]
Q3 = quartis[0.75]
print(f"Q1: {Q1}")
print(f"Q2 (Mediana): {Q2}")
print(f"Q3: {Q3}")

print("______________________________________________________")

desvio_padrao = round(df_tratado['Salario'].std(ddof=1), 2 )
print(desvio_padrao)

print("______________________________________________________")


media = df_tratado['Salario'].mean()
coef_var = (desvio_padrao / media)*100%
print(coef_var)



```
