## 1) Carrega o dataset para um DataFrame (df) e inspeciona as primeiras linhas.

```python
import pandas as pd


df = pd.read_csv("PL06_Dados_Analise_Dataset.csv")

df.head()
```

## 2) Para a coluna `Leads`: Média, Mediana, Moda, Quartis (Q1, Q2, Q3), Desvio padrão e Coeficiente de variação.

```python
col = 'Leads'
s = df[col]

media = s.mean()
mediana = s.median()
moda = s.mode()            # pode ter 1 ou mais valores
q1 = s.quantile(0.25)
q2 = s.quantile(0.50)      
q3 = s.quantile(0.75)
desvio_padrao = s.std()  # amostral (ddof=1 por omissão)
coef_var = desvio_padrao / media

print('--- Estatísticas: Leads ---')
print('Média:', media)
print('Mediana:', mediana)
print('Moda(s) :', moda)
print('Q1 (25%):', q1)
print('Q2 (50%) = Mediana:', q2)
print('Q3 (75%):', q3)
print('Desvio padrão (amostral):', desvio_padrao)
print('Coeficiente variação:', coef_var)
##
##Quanto maior o valor do CV, maior é a variabilidade dos dados em relação à média, indicando mais heterogeneidade ou instabilidade.

##Quanto menor o CV, mais homogêneos e consistentes são os dados, ou seja, menor a dispersão relativa.
```

## 3) Repete os mesmos cálculos para a coluna `Vendas`.

```python
col = 'Vendas'
s = df[col]

media_v = s.mean()
mediana_v = s.median()
moda_v = s.mode()
q1_v = s.quantile(0.25)
q2_v = s.quantile(0.50)
q3_v = s.quantile(0.75)
desvio_padrao_v = s.std()
coef_var_v = desvio_padrao_v / media_v

print('--- Estatísticas: Vendas ---')
print('Média:', media_v)
print('Mediana:', mediana_v)
print('Moda:', moda_v)
print('Q1 (25%) :', q1_v)
print('Q2 (50%) = Mediana:', q2_v)
print('Q3 (75%):', q3_v)
print('Desvio padrão (amostral):', desvio_padrao_v)
print('Coeficiente variação:', coef_var_v)
```

## 4) Interpreta: a média e a mediana estão próximas? Há uma moda bem definida?

```python
# Interpretação simples para Leads
dif_leads = abs(media - mediana)
if dif_leads <= 1:
    prox_leads = 'próximas'
else:
    prox_leads = 'afastadas'

if len(moda) == 1:
    moda_txt_leads = 'Unimodal.'
elif len(moda) > 1:
    moda_txt_leads = 'Multimodal (várias modas).'
else:
    moda_txt_leads = 'Amodal.'

print('LEADS -> Média e mediana estão', prox_leads, '(diferença ≈', dif_leads, ').', moda_txt_leads)

# Interpretação simples para Vendas
dif_vendas = abs(media_v - mediana_v)
if dif_vendas <= 1:
    prox_vendas = 'próximas'
else:
    prox_vendas = 'afastadas'

if len(moda_v) == 1:
    moda_txt_vendas = 'Unimodal'
elif len(moda_v) > 1:
    moda_txt_vendas = 'Multimodal (várias modas).'
else:
    moda_txt_vendas = 'Amodal'

print('VENDAS -> Média e mediana estão', prox_vendas, '(diferença ≈', dif_vendas, ').', moda_txt_vendas)
```

## 5) Correlação de Pearson entre `Leads` e `Vendas`. Interpreta o sinal e a magnitude.

```python
r_lv = df['Leads'].corr(df['Vendas'])

# Classificação simples da magnitude
a = abs(r_lv)
if a < 0.09:
    mag = 'muito fraca'
elif a < 0.29:
    mag = 'fraca'
elif a < 0.49:
    mag = 'moderada'
elif a < 0.69:
    mag = 'forte'
else:
    mag = 'muito forte'

if r_lv > 0:
    sinal = "positivo"
elif r_lv < 0:
    sinal = "negativo"
else:
    sinal = "nulo"


print('Correlação Pearson (Leads, Vendas):', r_lv, '->', mag, '; sinal', sinal)
```

## 6) Correlação entre `Vendas` e `Horas_Formacao` e comentário de um possível efeito.

```python
r_vh = df['Vendas'].corr(df['Horas_Formacao'])

a = abs(r_vh)
if a < 0.09:
    mag = 'muito fraca'
elif a < 0.29:
    mag = 'fraca'
elif a < 0.49:
    mag = 'moderada'
elif a < 0.69:
    mag = 'forte'
else:
    mag = 'muito forte'

if r_vh > 0:
    sinal = "positivo"
elif r_vh < 0:
    sinal = "negativo"
else:
    sinal = "nulo"
    
print('Correlação Pearson (Vendas, Horas_Formacao):', r_vh, '->', mag, '; sinal', sinal)

if r_vh > 0:
    print('Comentário: Associação positiva — períodos com mais formação podem coincidir com ligeiro aumento nas vendas (não implica causalidade).')
elif r_vh < 0:
    print('Comentário: Associação negativa — mais formação pode coincidir com menor foco comercial nesse período (ou outras variáveis).')
else:
    print('Comentário: Relação linear praticamente nula.')
```
