````python
import matplotlib.pyplot as plt
import numpy as np


idade_clientes = np.random.normal(loc=35, scale=10, size=1000)
unidades_vendidas = [150, 220, 180, 250, 190]
produto = ['A', 'B', 'C', 'D', 'E']

visitas_site = [10, 25, 45, 50, 70, 80, 95, 110, 120, 150]
valor_gasto = [50, 80, 120, 150, 200, 250, 280, 320, 350, 400]

# Criar painel de 2x2


# Quadro 1 (Superior Esquerda) - Histograma
idade_clientes = np.random.normal(loc=35, scale=10, size=1000)
plt.subplot(2, 2, 1) 
plt.hist(idade_clientes, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title("Distribuição de Idade dos Clientes")


# Quadro 2 (Superior Direita) - Gráfico de Barras

unidades_vendidas = [150, 220, 180, 250, 190] ## axis = y
produto = ['A', 'B', 'C', 'D', 'E'] ## axis = x

plt.subplot(2, 2, 2) 
plt.bar(produto, unidades_vendidas, color='orange')
plt.title("Popularidade dos Produtos")
plt.xlabel("Produto")
plt.ylabel("Unidades Vendidas")

# Quadro 3 (Inferior Esquerda) - Gráfico de Setores

vendas_regiao = [40, 30, 15, 15]
regioes = ['Norte', 'Sul', 'Centro', 'Ilhas']

plt.subplot(2, 2, 3) 
plt.pie(vendas_regiao, labels=regioes, autopct="%1.1f%%", startangle=90,colors=['blue', 'green', 'orange', 'purple'])
plt.title("Vendas por Região")

# Quadro 4 (Inferior Direita) - Gráfico de Dispersão
plt.subplot(2, 2, 4) 
plt.scatter(visitas_site, valor_gasto, color='purple', marker='o')
plt.title("Relação Visitas vs. Valor Gasto")
plt.xlabel
plt.ylabel("Valor Gasto (€)")

# Ajustar espaçamento
plt.tight_layout()
plt.show()
````
