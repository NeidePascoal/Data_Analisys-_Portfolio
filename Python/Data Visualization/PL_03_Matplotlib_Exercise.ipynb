```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Gráfico de Barras
import matplotlib.pyplot as plt

x1 = [0.25, 2.25, 3.25, 5.25, 7.25]
y1 = [300, 400, 200, 600, 700]
x2 = [0.75, 1.75, 2.75, 3.75, 4.75]
y2 = [50, 30, 20, 50, 60]

# cria um grafico de barra para o carpinteiro
plt.bar (x1, y1, label = 'carpinteiros', color='b', width=0.5)

# canalizador
plt.bar (x2, y2, label = 'canalizador', color='g', width=0.5)

# adiciona linhas de grades horizontais 
#axis= y cria apenas linhas orizontais no eixo y
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.legend()
plt.xlabel('Dias')
plt.ylabel('Salário')
plt.title('Gráficode Barras Verticais')
plt.show()
```

```python
## gráfico de barras horizontais -- BARH

x1 = [0.25, 2.25, 3.25, 5.25, 7.25]
y1 = [300, 400, 200, 600, 700]
x2 = [0.75, 1.75, 2.75, 3.75, 4.75]
y2 = [50, 30, 20, 50, 60]

# cria um grafico de barra para o carpinteiro
plt.barh (x1, y1, label = 'carpinteiros', color='b', height=0.5)

# canalizador
plt.barh (x2, y2, label = 'canalizador', color='g', height=0.5)

# adiciona linhas de grades horizontais 
#axis= y cria apenas linhas orizontais no eixo y
plt.grid(axis='x', linestyle='--', alpha=0.6)

### Os eixos estão invertidos 
plt.legend()
plt.ylabel('Dias')
plt.xlabel('Salário')
plt.title('Gráficode Barras Verticais')
plt.show()
```

```python
### gráficode Barras empilhadas 
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas_produtoA =[ 150, 180, 200, 220]
vendas_produtoB =[80, 100, 120, 150]

# O PLT.BAR O PROD. A DEFINE A BASE DO GRAFICO

plt.bar(meses, vendas_produtoA, label= 'Produto A', color='orange')

plt.bar(meses, vendas_produtoB, bottom= vendas_produtoA, label='Produto B', color='skyblue')

plt.legend()
plt.xlabel('Meses')
plt.ylabel('Total de Vendas')
plt.title('Gráfico barras empilhadas')
plt.show()
```

```python
## Gráfico de barras com barras de erro---
equipas = ['equipa A', 'equipa B', 'equipa C']
pontuacoes_medias = [85, 92, 78]

desvio_padrão = [5,2,7]

plt.bar(equipas, pontuacoes_medias, yerr=desvio_padrão, capsize =5, color= 'teal')

plt.xlabel('Equipas')
plt.ylabel('Pontuação Média')
plt.title('Pontuação Média das Equipas com DV')
plt.show()

```

```python
### Barras agrupadas 
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas_produtoA =[ 150, 180, 200, 220]
vendas_produtoB =[80, 100, 120, 150]

x_indices = np.arange(len(meses))
largura = 0.35
plt.bar(x_indices-largura/2, vendas_produtoA, largura, label ='Produto A')
plt.bar(x_indices+largura/2, vendas_produtoB, largura, label ='Produto B')


plt.xticks(x_indices, meses)

plt.legend()
plt.xlabel('Meses')
plt.ylabel('Total de Vendas')
plt.title('Comparação de Vendas de Produtos por Mês')
plt.show()
```

```python
#### Gráfioc Histograma

dados = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(dados, bins=30, edgecolor='black', color='skyblue')

plt.xlabel('Intervalos de Notas')
plt.ylabel('Número de Alunos')
plt.title('Distribuição de Notas dos Alunos')
plt.show()
```

```python
## Gráfico histograma - comparar distribuições 
grupo_A = np.random.normal(loc=75, scale=5, size=500)
grupo_B = np.random.normal(loc=80, scale=8, size=500)

plt.hist(grupo_A, bins=20, alpha=0.7, label='Grupo A', color='lightgreen')

plt.hist(grupo_B, bins=20, alpha=0.7, label='Grupo B', color='coral')

plt.title('Comparação da Distribuiçao de Resultados entre dois grupos')
plt.xlabel('Resultados')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
### Grafico de dispersão 
grupo_A_notas = np.random.normal(loc=70, scale=8, size=100)
grupo_B_notas = np.random.normal(loc=85, scale=6, size=100)

## gerar dados correlacionados 
horas_estudo_A = np.random.normal(loc=5, scale=1, size=100)
horas_estudo_B = np.random.normal(loc=8, scale=1, size=100)

plt.scatter(horas_estudo_A, grupo_A_notas, color='skyblue', label='grupo_A', alpha=0.7)


plt.scatter(horas_estudo_B, grupo_B_notas, color='coral', label='grupo_B', alpha=0.7)

plt.title('Notas de Exame vs Horas de Estudo por Grupo')
plt.xlabel('Hosras de Estudo')
plt.ylabel('Notas dos Exames')
plt.grid(True, linestyle = '--', alpha=0.6)
plt.legend()
plt.show()

```

```python
##Gráficode Dispersão __ scatter

horas_estudo = np.array([2,3,5,6,8,9,10,11])
nota_exame = np.array([55, 65, 70, 75, 85, 90, 95, 98])

tamanho_turma = np.array([100, 150, 200, 250, 300, 350, 400, 450])
cor_desempenho = np.array([1,2,3,4,5,6,7,8,])

plt.scatter( horas_estudo, nota_exame, s= tamanho_turma, c= cor_desempenho, cmap='viridis')

plt.title(' Horas de Estudoe Nota Exame')
plt.xlabel('Hosras de Estudo diárias')
plt.ylabel('Notas do Exame')
plt.grid(True, linestyle = '--', alpha=0.6)
plt.colorbar(label='Indice de desemprenho')
plt.show()
```

```python
slice= [12, 25, 50, 36, 19]
activites= ['NLP','Neural Network', 'Data analytics', 'Quantun Computing', 'Machine Learnig']

colors =['r', 'b', 'c', 'g', 'orange']
plt.pie(slice,
        labels= colors,
        startangle= 90,
        shadow= True,
        explode= (0,0.1,0,0,0),
        autopct= '%1.1f%%')
plt.title('Temas de Treino')
plt.show()
```

```python
### gráfico de sectores pie(wedgepropds=)
categorias =['Marketing', 'Rcursos Humanos', 'Operações', 'Investigação e Desenvolvimento']

despesas = [30, 20, 40, 10]

destaque= [0,0,0.1,0] # destacar a fatia " operações "

plt.pie(despesas,
        labels= categorias,
        explode=destaque,
        autopct='%1.1f%%',
        shadow= True,
        startangle=90,
        wedgeprops= dict(width=0.4, edgecolor='w')
        )
plt.title('Distribuição de Despesas Mensais')
plt.show ()
```

```python
## Gráfico de Caixa 0 boxplot

np.random.seed (42) ## define a semente para gerar os dados aleatorios 

notas_turma_A = np.random.normal( loc=70, scale=8, size= 50)

notas_turma_B = np.random.normal( loc=85, scale=6, size= 50)

notas_turma_C = np.random.normal( loc=78, scale=12, size= 50)

dados_para_plotar =[notas_turma_A, notas_turma_B, notas_turma_C]

nomes_turmas = ['TurmaA', 'Turma B', 'Turma C']

plt.boxplot(dados_para_plotar, labels= nomes_turmas)

plt.title(' Comparação da Distribuição de Notas entre Turmas ')
plt.xlabel('Hoara de Estudo')
plt.ylabel('Notas')
plt.show()
```
