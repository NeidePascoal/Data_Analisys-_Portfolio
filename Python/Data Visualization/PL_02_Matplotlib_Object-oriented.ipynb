# Método Orientado a Objeto Matplotlib¶


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# %matplotlib inline: É o comando mágico para exibir os gráficos do Matplotlib dentro do notebook, logo abaixo da célula de código.
# Sem ele, o gráfico poderia aparecer numa janela separada ou não ser exibido de todo.
# É uma boa prática incluí-lo, pois embora o Jupyter já o faça automaticamente, garante que o teu código irá funcionar da mesma forma se um dia o executares num script Python tradicional (.py).
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
     ax.plot(x, y, 'g')
     ax.set_xlabel('x')
     ax.set_ylabel('y')
     ax.set_title('título')

plt.tight_layout()
plt.show() # O plt.show() é necessário para exibir a figura em alguns ambientes, como scripts Python. Em jupyter Notebooks, a figura geralmente é exibida automaticamente ou chamando apenas o fig. Então, podes optar por usar plt.show() ou apenas fig, mas o o uso do show() é uma prática comum para garantir que a figura seja exibida corretamente em todos os ambientes.
```

### Legendas, rótulos e títulos
Agora que abordamos os fundamentos de como criar uma tela de figura e adicionar instâncias de eixos à tela, vamos ver como decorar uma figura com títulos, rótulos de eixo e legendas.

#### Títulos das figuras

Um título pode ser adicionado a cada instância de eixo numa figura. Para definir o título, use o método set_title na instância de axes:

```python
ax.set_title("título")
```

##### Rótulos de eixo

Da mesma forma, com os métodos set_xlabel e set_ylabel, podemos definir os rótulos dos eixos X e Y:

```python
ax.set_xlabel("x")
ax.set_ylabel("y")
```

#### Legendas
Pode usar o argumento label="label text" quando gráficos ou outros objetos são adicionados à figura e, em seguida, usar o método legend sem argumentos para adicionar a legenda à figura:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="Legenda 1")
ax.plot(x, x**3, label="Legenda 2")
ax.legend()
```

Observe que como as legendas se sobrepõem ao plot.

A função legend aceita um argumento opcional loc que pode ser usado para especificar onde na figura a legenda deve ser desenhada. Os valores permitidos de loc são códigos numéricos para os vários lugares onde a legenda pode ser desenhada. Consulte a página de documentação para obter detalhes. Alguns dos valores loc mais comuns são:

```python
# Muitas opções....

ax.legend(loc=1) # canto superior direito
ax.legend(loc=2) # canto superior esquerdo
ax.legend(loc=3) # canto inferior esquerdo
ax.legend(loc=4) # canto inferior direito

# .. muitas outras opções estão disponíveis

# Mais comum para escolher
ax.legend(loc=0) # permite que matplotlib decida a localização ideal.
fig
```

### Definir cores, larguras de linha, tipos de linha
O Matplotlib oferece muitas opções para personalizar cores, larguras de linha e tipos de linha.

Existe a sintaxe básica do MATLAB (não recomendado):

#### Cores com a syntax de MatLab
Com o matplotlib, podemos definir as cores das linhas e outros elementos gráficos de várias maneiras. Em primeiro lugar, podemos usar a sintaxe do MATLAB, onde 'b' significa azul, 'g' significa verde, etc. A API MATLAB para selecionar estilos de linha também é suportada: onde, por exemplo, 'b.-' significa um linha azul com pontos:

```python
# Cor de linha e estilo com a syntax do MATLAB
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # linha azul com pontos
ax.plot(x, x**3, 'g--') # linha tracejada verde
```

### Cores com o color=parameter
Também podemos definir cores pelos nomes ou códigos hexadecimais RGB e, opcionalmente, fornecer um valor alfa usando os argumentos 'color' e 'alpha'. Alpha indica a opacidade.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # meio transparente
ax.plot(x, x+2, color="#DC62DC") # Código hexadecimal RGB
ax.plot(x, x+3, color="#C69E6D") # Código hexadecimal RGB
```

#### Estilos de linha e marcadores
Para alterar a largura da linha, podemos usar o argumento 'linewidth' ou 'lw'. O estilo de linha pode ser selecionado usando os argumentos de 'linestyle' ou 'ls':

```python
fig, ax = plt.subplots(figsize=(12,6))   # tamanho da figura em polegadas (width, height) por 
ax.plot(x, x+1, color="red", linewidth=0.25)# largura da linha em pontos
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possíveis opções de tipo de linha ‘-‘, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-') # lw é abreviação de linewidth 
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# traço personalizado
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10])# formato: comprimento da linha, comprimento do espaço. Ou seja, 5 pontos, 10 pontos em branco, 15 pontos, 10 pontos em branco

# símbolos de marcadores possíveis: marcador = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+') # marcador de mais será desenhado em cada ponto de dados
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o') # x+10 terá marcadores circulares
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# tamanho e cor do marcador
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2) # ax.plot está a desenhar círculos pequenos em cada ponto de dados
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Controlo sobre a aparência do eixo
Nesta seção, veremos como controlar as propriedades de dimensionamento do eixo numa figura matplotlib.

#### Plot range
Podemos configurar os intervalos dos eixos usando os métodos 'set_ylim' e 'set_xlim' no objeto de eixo, ou axis('tight') para obter automaticamente intervalos de eixos "bem ajustados":

```python
# Uso similar a plt.figure() exceto usar descompactação de tuplo para apanhar a figura e os eixos
# Cria uma figura e uma matriz (array) de 1 linha por 3 colunas de objetos 'Axes'.
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4)) # tamanho da figura em polegadas (width, height) e especificamos 3 gráficos lado a lado

# --- Manipular o array de Axes ---
# Acedemos ao primeiro gráfico (índice 0) do array 'axes'.
# Este gráfico usa os limites de eixo automáticos (default). O Matplotlib adiciona automaticamente margens para um melhor visual.
axes[0].plot(x, x**2, x, x**3) # desenha duas linhas no primeiro gráfico 
axes[0].set_title("Intervalos de Eixos Default")

# Acedemos ao segundo gráfico (índice 1).
# O comando axis("tight") remove todas as margens, ajustando o gráfico apenas aos dados.
axes[1].plot(x, x**2, x, x**3)
axes[1].axis("tight")
axes[1].set_title("Eixos Tight")

# Acedemos ao terceiro e último gráfico (índice 2).
# Definimos os limites dos eixos manualmente para um controlo total do zoom.
axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Intervalo de Eixos Personalizados")
```
