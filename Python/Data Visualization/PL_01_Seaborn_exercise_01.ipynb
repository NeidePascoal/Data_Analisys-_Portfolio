#### Gráfico de Dispersão

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("notebook")  # este é o default
# Define o estilo e a paleta para um visual mais limpo
sns.set_style('whitegrid')
sns.set_palette('deep')
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')

# Cria o gráfico de dispersão com múltiplas variáveis
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',      # Cor por género. deve 
    style='smoker',   # Estilo do marcador por fumador
    size='size'       # Tamanho do ponto por tamanho do grupo, deve ser 
)
plt.title('Gorjeta por Conta Total, Sexo, Fumador e Tamanho do Grupo')
plt.xlabel('Conta Total (€)')
plt.ylabel('Gorjeta (€)')

# Coloca a legenda fora do gráfico para não sobrepor os pontos
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

#### Gráfico de Linhas

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')
# Define o estilo e a paleta
sns.set_style('whitegrid') # whitegrid é o estilo do grafico
sns.set_palette('rainbow') # rocket é a paleta de cores

# Cria o gráfico de linhas
sns.lineplot(
    data=tips,
    x='day',
    y='tip',
    hue='smoker',
    # 'hue' separa as linhas pela variável 'smoker' o que queremos analisar
    # 'style' altera o estilo de linha pela variável 'smoker'
    markers=True # 'markers' adiciona um marcador em cada ponto de dados
)
plt.title('Gorjeta Média por Dia e por Fumador')
plt.xlabel('Dia da Semana')
plt.ylabel('Gorjeta Média ($)')
plt.show()

# o sombreado do grafico indica o intervalo de confiança (confidence interval - CI) em relação à média. os pontos representam a média das gorjetas por dia, e o sombreado ao redor da linha indica a variação esperada (intervalo de confiança) dessas médias.
```

# Tipos de Gráficos - **Gráficos de Distribuição**

#### Gráfico histograma

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')
# Define o estilo e a paleta
sns.set_style('whitegrid')
sns.set_palette('viridis')

# Cria um histograma para a variável 'total_bill'
# O argumento 'hue' divide o histograma por uma variável categórica
sns.histplot(
    data=tips,
    x='total_bill',
    bins=20, # número de barras
    hue='smoker',   # analisa a variável 'smoker', deve ser uma variável categórica
    kde=True #Kernel Density Estimation (KDE) é um método não-paramétrico 
# para estimar suavemente a função densidade de probabilidade de uma amostra.
)
plt.title('Distribuição do Valor da Conta Total por Hábito de Fumar')
plt.xlabel('Conta Total (€)')
plt.ylabel('Frequência')
plt.show()
```

#### Gráfico de Densidade

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')
# Define o estilo e a paleta
sns.set_style('whitegrid')
sns.set_palette('Paired')

# Cria um kdeplot para a variável 'total_bill' com argumentos adicionais
sns.kdeplot(
    data=tips,
    x='total_bill',
    hue='sex',
    fill=True, ## preenche o gráfico
    common_norm=False, # normaliza cada curva individualmente
    alpha=0.6 # controla a transparência do preenchimento
)
plt.title('Distribuição da Conta Total por Sexo')
plt.xlabel('Conta Total (€)')
plt.show()
```

#### Gráfico de Barras

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')
# Define o estilo
sns.set_style('whitegrid')

# Cria o gráfico de barras
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
# Podemos incluir o argumento 'estimator', que calcula o valor de cada barra.
# Pode ser usado para calcular a média(padrão do seaborn), soma, mediana.

plt.title('Média da Conta por Dia e por Sexo')
plt.xlabel('Dia da Semana')
plt.ylabel('Média da Conta Total (€)')
plt.show()

# neste grafico, cada barra representa a média da conta total para homens 
# e mulheres em cada dia da semana. As barras azuis correspondem
# aos homens, enquanto as barras azuis escuras correspondem às mulheres.
# Podemos observar que, em geral, os homens tendem a ter uma média de conta
# total ligeiramente mais alta do que as mulheres em todos os dias da semana.
```

#### Gráfico de Caixas

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')
# Define o estilo
sns.set_style('whitegrid')

# Cria o gráfico de caixas
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')

plt.title('Distribuição da Conta por Dia e por Sexo')
plt.xlabel('Dia da Semana')
plt.ylabel('Conta Total (€)')
plt.show()


```

#### Gráfico de Contagem

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')
# Define o estilo e a paleta
sns.set_style('whitegrid')
sns.set_palette('rocket')

# Encontra a ordem das categorias do dia com base na contagem
order_dias = tips['day'].value_counts().index

# Cria o gráfico de contagem com a ordem definida
sns.countplot(
    data=tips,
    x='day',
    hue='sex', # hue variaveel que queremos analisar e deve ser categórica não numérica
    order=order_dias
    
)
plt.title('Número de Clientes por Dia e por Sexo')
plt.xlabel('Dia da Semana')
plt.ylabel('Contagem de Clientes')
plt.show()
```

```python
# Exemplo de um gráfico de contagem horizontal
sns.countplot(data=tips, y='sex', orient='h')
plt.title('Contagem de Clientes por Sexo (Gráfico Horizontal)')
plt.show()
```

#### Gráfico Conjunto

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')

# Cria um jointplot com um gráfico de regressão no centro
# O 'kind' pode ser 'scatter', 'hex', 'kde' ou 'reg'
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg') ## kinf = tipo de grafico, 

plt.suptitle('Análise Conjunta da Conta e Gorjeta', y=1.02)
# com este gráfico o comando plt.title() não funciona.
plt.show()
```

#### Gráfico de Pares

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Carrega o conjunto de dados 'tips'
tips = sns.load_dataset('tips')

# Cria o gráfico de pares
# 'hue' adiciona uma camada de análise, colorindo os pontos por género
sns.pairplot(data=tips, hue='sex')

plt.suptitle('Relação entre as Variáveis Numéricas por Sexo', y=1.02)
plt.show()
```
