## Método orientado a objeto- Matplolib

```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
```

```python
x= np.linspace(0, 5, 11) # cria um array com 11 pontos entre 0 e 5
y = x**2    # cria um array com os valores de x ao quadrado x^2 (raiz quadrada de x )
```

```python
fig= plt.figure()  # cria a figura

# vamos adicionar um unico gráfico à figura
axes =fig.add_axes([0.1, 0.1, 0.8, 0.8]) ## [ representa a posição do gráfico na figura] --[esquerda, baixo, largura, altura]

## vamos plotar os dados x e y
axes.plot(x, y, 'b')  # plota x e y com linha azul
axes.set_xlabel('Eixo X')  # nomeia o eixo x
axes.set_ylabel('Eixo Y')  # nomeia o eixo y
axes.set_title('Gráfico X vs Y')  # nomeia o gráfico
plt.show()  # mostra o gráfico

```

```python
## subposição de graficos 

fig= plt.figure()  # cria a figura
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # cria o primeiro gráfico
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # cria o segundo gráfico # [ esquerda, baixo, largura, altura]  ocupa 40% da largura e 30% da altura da figura

# vamos plotar os dados x e y no primeiro gráfico
axes1.plot(x, y, 'b')  # plota x e y com linha azul
axes1.set_xlabel('Eixo X')  # nomeia o eixo x       
axes1.set_ylabel('Eixo Y')  # nomeia o eixo y
axes1.set_title('Gráfico X vs Y')  # nomeia o gráfico       

# vamos plotar os dados x e y no segundo gráfico
axes2.plot(y, x, 'r')  # plota x e y com linha vermelha
axes2.set_xlabel('Eixo X')  # nomeia o eixo x   
axes2.set_ylabel('Eixo Y')  # nomeia o eixo y
axes2.set_title('Gráfico X vs Y')  # nomeia o gráfico       


plt.show()  # mostra o gráfico
```

```python
# uso similar do plt.figure e plt.subplot
fig, axes=plt.subplots(nrows=1, ncols=1)  # cria uma figura com 1 linha e 2 colunas

axes.plot(x, y, 'b')  # plota x e y com linha azul no primeiro gráfico       
axes.set_xlabel('Eixo X')  # nomeia o eixo x       
axes.set_ylabel('Eixo Y')  # nomeia o eixo y 
axes.set_title('Gráfico X vs Y')  # nomeia o gráfico

plt.tight_layout()  # ajusta o layout para evitar sobreposição
plt.show()  # mostra o gráfico
```

```python
## vamos criar uma figura com 1 linha e 2 colunas   

fig, axes = plt.subplots(nrows=1, ncols=2, ) # figsize=(10,5)  # cria uma figura com 1 linha e 2 colunas, tamanho 10x5 polegadas

#Vamos criar um loop que itera sobr e cada objeto axes individualmente      

for ax in axes:  # itera sobre cada gráfico -- iterar significa passar por cada um dos gráficos
    ax.plot(x,y, 'b')  # plota x e y com linha azul em cada gráfico
    ax.set_xlabel('Eixo X')  # nomeia o eixo x em cada gráfico
    ax.set_ylabel('Eixo Y')  # nomeia o eixo y em cada gráfico
    ax.set_title('Gráfico X vs Y')  # nomeia o gráfico em cada gráfico
```

```python
### tamanha da figura DPI 
import numpy as np
fig = plt.figure(figsize=(8,4), dpi=100)  # cria uma figura com tamanho 8x4 polegadas e resolução de 100 pontos por polegada
```

```python


fig, axes = plt.subplots(figsize=(12,4), dpi=100)  # cria uma figura com 2 linhas e 1 coluna, tamanho 8x4 polegadas e resolução de 100 pontos por polegada

axes.plot(x, y, 'b')  # plota x e y com linha azul no primeiro gráfico
axes.set_xlabel('Eixo X')  # nomeia o eixo x
axes.set_ylabel('Eixo Y')  # nomeia o eixo y        
axes.set_title('Gráfico X vs Y')  # nomeia o gráfico

plt.tight_layout()  # ajusta o layout para evitar sobreposição
plt.show()  # mostra o gráfico


```
