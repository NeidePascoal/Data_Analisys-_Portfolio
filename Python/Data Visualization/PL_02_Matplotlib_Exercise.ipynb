````python
import matplotlib.pyplot as plt

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(x, popularity, color='blue', width=0.5)

plt.minorticks_on()

# Grid menor (escuro, mais frequente)
plt.grid(which='minor', axis='both', color='gray', linestyle='-', linewidth=0.5, alpha=0.5)

# Grid maior (vermelho, menos frequente)
plt.grid(which='major', axis='both', color='red', linestyle='-', linewidth=0.5, alpha=0.7)

plt.xlabel('Linguagens')
plt.ylabel('Popularidade')
plt.title('Popularidade das linguagens de programação')
plt.show()
````

````python
python
import matplotlib.pyplot as plt
import numpy as np

notas_teste_1 = np.random.randint(60, 100, 50)
notas_teste_2 = np.random.randint(50, 95, 50)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Histograma do Teste 1
axs[0].hist(notas_teste_1, bins=8, color='blue')
axs[0].set_title('Notas do Teste 1')
axs[0].set_xlabel('Notas')
axs[0].set_ylabel('Frequência')
axs[0].grid(True)

# Histograma comparativo Teste 1 x Teste 2
axs[1].hist(notas_teste_1, bins=8, alpha=0.7, label='Teste 1', color='blue')
axs[1].hist(notas_teste_2, bins=8, alpha=0.5, label='Teste 2', color='red')
axs[1].set_title('Comparação de Notas')
axs[1].set_xlabel('Notas')
axs[1].set_ylabel('Frequência')
axs[1].legend()
axs[1].grid(True)
````
